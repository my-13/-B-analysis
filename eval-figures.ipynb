```python
markersize= 13
IMU_keys = IMU.keys()
EMG_keys = EMG.keys()


x = np.linspace(0,len(IMU_val)-1,64)# x coordinates at which to evaluate interpolated values
xp = np.linspace(0,len(IMU_val)-1,len(IMU_val))# x coordinates of data points
IMU_resampled = IMU.apply(lambda value: np.interp(x,xp,value))#new_points

x = np.linspace(0,len(EMG_val)-1,64)# x coordinates at which to evaluate interpolated values
xp = np.linspace(0,len(EMG_val)-1,len(EMG_val))# x coordinates of data points
EMG_resampled = EMG.apply(lambda value: np.interp(x,xp,value))#new_points
xx = np.linspace(0,len(IMU_val)/148.148148148,64)

fig,axs = plt.subplots(1,2,figsize=(7,3),sharey=True,sharex=True)
linewidth=3
sns.lineplot(x =xx, y=IMU_resampled[IMU_keys[3]].values,ax=axs[0],label="IMU",color=palette[2],linewidth=linewidth,linestyle='solid')
sns.lineplot(x =xx, y=IMU_resampled[IMU_keys[40]].values,ax=axs[0],palette=palette,color=palette[2],linewidth=linewidth,linestyle='dashed')
sns.lineplot(x =xx, y=EMG_resampled[EMG_keys[10]].values,ax=axs[0],label="EMG",color=palette[0],linewidth=linewidth,linestyle='solid')
sns.lineplot(x =xx, y=EMG_resampled[EMG_keys[6]].values,ax=axs[0],palette=palette,color=palette[0],linewidth=linewidth,linestyle='dashed')
# ax.legend(bbox_to_anchor=(1.05, 1.4), loc='upper right',frameon=False,framealpha=1, ncol=3)
# ax.set_xlabel('Time (sec)')
# ax.set_ylabel("Magnitude")


EMG_normalized = EMG_resampled.apply(lambda val: val - np.mean(val))
stdev = EMG_normalized.values.std()
EMG_normalized = EMG_normalized.apply(lambda val: val / stdev)

IMU_normalized = IMU_resampled.apply(lambda val: val - np.mean(val))
stdev = IMU_normalized.values.std()
IMU_normalized = IMU_normalized.apply(lambda val: val / stdev)
sns.lineplot(x =xx, y=IMU_normalized[IMU_keys[3]].values,ax=axs[1],label="IMU signal",color=palette[2],linewidth=linewidth,linestyle='solid')
sns.lineplot(x =xx, y=IMU_normalized[IMU_keys[40]].values,ax=axs[1],palette=palette,label="_IMU noise",color=palette[2],linewidth=linewidth,linestyle='dashed')
sns.lineplot(x =xx, y=EMG_normalized[EMG_keys[10]].values,ax=axs[1],label="EMG signal",color=palette[0],linewidth=linewidth,linestyle='solid')
sns.lineplot(x =xx, y=EMG_normalized[EMG_keys[6]].values,ax=axs[1],palette=palette,label="_EMG noise",color=palette[0],linewidth=linewidth,linestyle='dashed')
fig.text(0.55, 0.01, 'Time (sec)', ha='center')
axs[0].set_ylabel("Biosignal Magnitude")
axs[0].set_title("Before Normalization")
axs[1].set_title("After Normalization")

for ix,ax in enumerate(axs):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_legend().remove()
# axs[0].legend(loc='upper right',frameon=False,framealpha=1, ncol=1)
axs[0].legend(frameon=False,framealpha=1)
plt.tight_layout()
# save plot
fname='normalize'
fig = plt.gcf()
for fmt in fmts:
    if not(os.path.exists(fmt)): os.mkdir(fmt)
    fig.savefig(os.path.join(fmt,fname+"."+fmt),bbox_inches='tight',pad_inches=0,dpi=500)
```
